프레임 병합
- pd.merge(frm, frm, how) 
- 공통의 열을 기준으로 병합 how = inner, left, right, outer
- frm.join() : 인덱스를 기준으로 병합
- concat()   : 연결

```python
frm01 = pd.DataFrame({
    '학번' : [1,2,3,4],
    '이름' : ['김채아', '오세헌', '김하윤', '서한나'],
    '학년' : [5, 6, 9, 6]
})
display(frm01)

frm02 = pd.DataFrame({
    '학번' : [5,4,2,3],
    '학과' : ['건강', '건강', 'MIS', '무역'],
    '학점' : [3.9, 3.8, 3.7, 4.5]
})
display(frm02)
```

```python
pd.merge(frm01, frm02, how = 'inner') # inner는 공통의 열을 기준으로 교집합인 것들을 출력해준다.
```

```python
pd.merge(frm01, frm02, how = 'outer') # outer는 교집합에 들어가지 못하는 데이터까지 포함시킨다.
```

```python
pd.merge(frm01, frm02, how = 'left') # left는 차집합의 의미를 가진다.
```

```python
seoulCt.rename(columns = {'기관명' : '구별'},inplace=True)
```

```python
# print(seoulPp.columns)
seoulPp = pd.DataFrame({"구별" : seoulPp['자치구']
,"인구수" : seoulPp['인구']
,"한국인" : seoulPp['인구.3']
,"외국인" : seoulPp['인구.6']
,"고령자" : seoulPp['65세이상고령자']})
display(seoulPp)
```

```python
print('step02. - cctv와 인구 현황 데이터 파악 - ') 
# display(seoulCt)
# display(seoulPp)
print('소계를 기준으로 오름차순 정렬 - ')
seoulCt.sort_values(by = '소계', ascending = True, inplace=True)
seoulCt.reset_index(inplace = True)
seoulCt.drop('index',axis=1,inplace = True)
display(seoulCt)
print('cctv가 가장 많은 구 확인 - ')
seoulCt.iloc[-1]
```

```python
print('2014-2016년 cctv수를 더하고 2013년 이전 cctv 수로 나누어서 최근 3년간 cctv 증가율을 계산 - ')
plusCt = (seoulCt['2014년'].values+seoulCt['2015년'].values+seoulCt['2016년'].values)/seoulCt['2013년도 이전'].values
print('증가율이 가장 높은 구 확인 - ')
seoulCt['cctv 증가율'] = plusCt
seoulCt.sort_values('cctv 증가율', axis = 0, ascending = True, inplace = True)
seoulCt.iloc[-1]
display(seoulCt)
```

```python
print('step03 - 서울시 인구 데이터 파악')
#display(seoulPp.head())
print('drop - 합계 삭제')
seoulPp.drop(seoulPp.index[2], axis=0, inplace = True)
display(seoulPp)
print('구별 결측값 확인 - 만약, 결측값이 있다면 행 삭제')
seoulPp.dropna(inplace=True)
display(seoulPp)
```

```python
print('각 구별 전체 인구를 이용하여 구별 외국인비율과 고령자비율 계산하여 반영')
seoulPp.drop([0,1],inplace=True)
```

```python
foreignP = seoulPp['외국인'][:]/seoulPp['인구수'][:]
seoulPp['외국인비율'] = foreignP
olderP = seoulPp['고령자'][:]/seoulPp['인구수'][:]
seoulPp['고령자비율'] = olderP
print('인구수로 정렬하여 인사이트 찾아보자')
seoulPp.sort_values('인구수',inplace = True)
seoulPp
```
