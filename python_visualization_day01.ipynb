```python
def aryInfo(ary) : 
    print('type - ' , type(ary)) 
    print('shape - ' , ary.shape)
    print('ndim  - ' , ary.ndim)
    print('dtype - ' , ary.dtype)
    print()
    print('data  -')
    print(ary) 

def seriesInfo(s) :
    print('type   - ' , type(s)) 
    print('index  - ' , s.index) 
    print('values - ' , s.values)
    print('dtype  - ' , s.dtype)
    print()
    print('data  -')
    print(s) 

def frmInfo(frm) : 
    print('type    - ' , type(frm))
    print('shape   - ' , frm.shape)
    print('ndim    - ' , frm.ndim)
    print('row idx - ' , frm.index , type(frm.index))
    print('col idx - ' , frm.columns , type(frm.columns))
    print('values  - ' , type(frm.values))
    print(frm.values)
    print('data - ')
    display(frm)
```

```python
irisFrm = sns.load_dataset('iris')
irisFrm.head()
```

```python
print('bar chart 최종목적')
print('종을 기준으로 그룹화 - ')
subsetFrm = irisFrm.groupby('species')
print('피처명 변경 sepal_length	 꽃받침의 길이, etc')
print('피처명 변경 species -> target')
irisFrm.rename(columns = {'sepal_length' : '꽃받침의 길이',
                          'sepal_width' : '꽃받침의 너비',
                          'petal_length' : '꽃잎의 길이',
                          'petal_width' : '꽃잎의 너비',
                          'species' : 'target'},
               inplace = True)
```

```python
print('종을 기준으로 그룹화 - ')
subsetFrm = irisFrm.groupby('target')
print('그룹에 대한 각 피처의 평균을 구한다.')
display(subsetFrm.mean())
print('종별 평균을 막대그래프로 시각화')
```

```python
subsetFrm.mean().plot(kind = 'bar')

plt.xlabel('종', rotation = 360)
plt.xticks(rotation = 360)
plt.ylabel('수치',rotation = 360)
plt.legend(loc = 'best')
plt.show()
plt.close()
```
