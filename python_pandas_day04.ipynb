```python
def aryInfo(ary) : 
    print('type - ' , type(ary)) 
    print('shape - ' , ary.shape)
    print('ndim  - ' , ary.ndim)
    print('dtype - ' , ary.dtype)
    print()
    print('data  -')
    print(ary) 

def seriesInfo(s) :
    print('type   - ' , type(s)) 
    print('index  - ' , s.index) 
    print('values - ' , s.values)
    print('dtype  - ' , s.dtype)
    print()
    print('data  -')
    print(s) 

def frmInfo(frm) : 
    print('type    - ' , type(frm))
    print('shape   - ' , frm.shape)
    print('ndim    - ' , frm.ndim)
    print('row idx - ' , frm.index , type(frm.index))
    print('col idx - ' , frm.columns , type(frm.columns))
    print('values  - ' , type(frm.values))
    print(frm.values)
    print('data - ')
    display(frm)
```

merge : 열병합
- merge를 인덱스를 기준으로 병합
- how , left_on, right_on , left_index, right_index

```python
frm01 = pd.DataFrame({
    'col01' : ['busan','seoul','jeju']
    ,'col02' : [1,2,3]
    ,'col03' : [3,2,1]
})

display(frm01)
print()
frm02 = pd.DataFrame(data    = np.arange(12).reshape(6,2)
                    ,columns = ['feature01','feature02'])
display(frm02)
```

```python
frame01 = pd.DataFrame({
    'col01' : ['taiwan','china','korea']
    , 'col02' : [1,2,3]
    , 'col03' : [3,4,5]
})

display(frame01)
print()
frame02 = pd.DataFrame(data = np.arange(12).reshape(6,2)
                      ,columns = ['feature01', 'feature02'])
display(frame02)
```

```python
pd.merge(frm01, frm02, right_index = True, left_on = 'col02') # 인덱스를 기준으로 병합하게된다. (0,1,2) left right 모두 쓰는 것이 문법이다.
```

```python
pd.merge(frame01, frame02, right_index = True, left_on = 'col02')
```

 - join : 인덱스 병합의 프레임 전용 함수

```python
frm01.join(frm02, how = 'outer') # 인덱스에 대한 병합

# pd.merge(frm01, frm02, left_index = True, right_index = True) # 열병합
```

```python
frame01.join(frame02, how = 'left')
```

데이터를 분석하다보면 그룹에 통계량이 확인이 필요한 경우가 있다
- groupby(열 인덱스 | 인덱스)
- 데이터를 그룹으로 분할하여 독립된 그룹에 대해서 별도의 데이터 처리 또는 통계량
- split   : 특징에 따라 분류
- apply   : 함수를 적용(집계, 필터링, 변환)
- combine : 처리된 결과를 하나의 프레임으로 결합.
- agg(count, sum, mean, median, var, std, min, max....) 

```python
sampleFrm = pd.read_csv('./data/service_data_groupby_sample.csv',
           encoding = "cp949")

sampleFrm.info()
display(sampleFrm)
```

```python
print('Q1) 나이평균을 확인하고 싶다면 - ')
ageMean = np.mean(sampleFrm['age'])
```

```python
print('Q2) 지역별 나이평균을 확인하고 싶다면 - ')
# sampleFrm.groupby('region').get_group('경기') # region을 기준으로 그룹화한다.
# np.mean(sampleFrm.groupby('region').get_group('경기')['age'])
sampleFrm.groupby('region')['age'].mean() # groupby를 통해 바로 그룹별 특정 값을 인덱싱 할 수 있다.
```

```python
print('Q3) 성별을 기준으로 그룹을 나누고 싶다면 - ')
```

```python
tmp = sampleFrm.groupby('gender')[['height']].mean()
```

```python
tmp.reset_index(inplace=True)
tmp
```

```python
print('다중 통계량의 정보를 확인하고자 할 때 - agg()')
sampleFrm.groupby('gender').agg(['mean','var','count'])
```

```python
sampleFrm.groupby('gender')[ 'height' ].agg(**{'mean_height' : 'mean'}).reset_index() # 열인덱스를 바꾸기 위해선 **를 붙여줘야 한다.
```

```python
sampleFrm.groupby('gender')['age'].agg(['max', 'mean','min','median']).reset_index() # 여러개의 인덱스를 만들고 싶으면 열인덱스 변경이 불가하다.
```

```python
print('Q4) 성별에 따른 거주지의 최빈값을 확인하고 싶다면 - ')
sampleFrm.groupby('gender')['region'].agg(lambda x : x.mode()) #mode()로, 최빈값을 확인할 수 있다. lambda식을 사용해야한다.
```

```python
tmp = sampleFrm.groupby( ['gender','region'] )['height'].agg(**{'m' : 'mean'}) # get_group을 묶기 위해선 tuple 형식을 사용해야한다.
```

```python
tmp.index #  다중 그룹을 만들면 튜플형식으로 나타나게된다.
```

```python
print('Q5) 성별에 따른 키의 최소값, 최대값을 구한다면 - ')
```

```python
sampleFrm.groupby(['gender'])[['height','age']].agg(['min','max']).reset_index()
```

```python
print('Q1) 성별을 그룹으로해서 성별그룹의 나이 평균을 구하고, 이 값을 결측값에 대체 - ')
femaleAge = titanicFrm.groupby('sex')['age'].mean()['female']
maleAge = titanicFrm.groupby('sex')['age'].mean()['male']
```

```python
titanicFrm['age'] = tmp
titanicFrm['age'].isnull().sum()
```

```python
print('Q2) 성별을 그룹으로 나이를 내림차순 정렬하여 상위 5명씩 추출 - ')
```

```python
# subset = titanicFrm.sort_values('age',ascending = False).groupby('sex').head(5)
# display(subset)

# for key, group in titanicFrm.groupby('sex'):
#     print('key   - ', key)
#     print('group - ', group)

def top5(frm):
    return frm.sort_values(by='age', ascending = False).head(5)


titanicFrm.groupby('sex').apply(lambda frm : frm.sort_values(by='age', ascending = False).head(5))

titanicFrm.groupby('sex').apply(top5)
```
