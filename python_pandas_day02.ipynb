```python
def aryInfo(ary) :
    print('type -',type(ary))
    print('shape -',ary.shape)
    print('ndim -',ary.ndim)
    print('dtype -',ary.dtype)
    print()
    print('data -')
    print(ary)
    
def seriesInfo(s) :
    print('type - ',type(s))
    print('index -',s.index)
    print('value -',s.values)
    print('dtype -',s.dtype)
    print()
    print('data -')
    print(s)
```

```python
print('dict를 이용한 프레임 생성 - ')
frm = pd.DataFrame({
    'feature01' : [1,2,3], #feature01 은 열 인덱스
    'feature02' : [1,2,3], #feature02 은 열 인덱스
    'feature03' : [1,2,3]
}) # dict의 key가 index
print('type    - ', type(frm)) # type은 DataFrame
print('shape   - ', frm.shape) # 3행 3열
print('ndim    - ', frm.ndim)  # 2차원 (표형식이기 때문에)
print('row idx - ', frm.index   , type(frm.index))# RangeIndex(0,3,1)
print('col idx - ', frm.columns , type(frm.columns)) # feature01, feature02, feature03
print('values -')
print(frm.values) # 배열 형식 data만 꺼내게되면 배열이다.
print('data - ')
display(frm) # display하게 되면 표처럼 보인다.
```

```python
def frmInfo(frm) :
    print('type    - ', type(frm)) 
    print('shape   - ', frm.shape)
    print('ndim    - ', frm.ndim)
    print('row idx - ', frm.index   , type(frm.index))
    print('col idx - ', frm.columns , type(frm.columns))
    print('values -')
    print(frm.values) 
    print('data - ')
    display(frm) 
```

```python
titanicFrm = sns.load_dataset('titanic')
print('type - ',type(titanicFrm))
titanicFrm.head() # head는 상위 건을 보여주는 함수 default = 5
```

```python
print('선실등급의 인원 수를 확인하고 싶다면 - ')

titanicFrm['pclass'].value_counts().values # 시리즈다. / value_counts를 쓰게되면 선실등급별 인원수를 알 수 있다.
```

```python
print('프레임의 컬럼명을 확인하고 싶다면 - ')
titanicFrm.columns # 컬럼명을 확인할 수 있다.
```

```python
print('기존 column의 값을 활용해서 추가한 값을 파생컬럼으로 부른다.')
print("Q1) age_by_10 컬럼을 추가하고 싶다(기존나이에 10살을 더한 값을 컬럼으로 추가)")
titanicFrm['age_by_10'] = ((titanicFrm['age'] + 10).values).astype('int')
display(titanicFrm.head())
```

```python
print("Q2) parch와 sibsp 더한값에 1을 추가해서 family_no 컬럼 추가")
titanicFrm['family_no'] = titanicFrm['parch'] + titanicFrm['sibsp'] + 1
display(titanicFrm.head(10))
```

```python
print('Q3) 방금 만든 2개의 컬럼을 삭제하고 싶다면 - ')
print('drop( , axis = 1)')
titanicFrm.drop(['age_by_10','family_no'],axis=1, inplace = True)
```

```python
maxFare  = np.max(titanicFrm['fare'])
minFare  = np.min(titanicFrm['fare'])
meanFare = np.mean(titanicFrm['fare'])
sumFare  = np.sum(titanicFrm['fare'])

print(maxFare)
print(minFare)
print(meanFare)
print(sumFare)
```
